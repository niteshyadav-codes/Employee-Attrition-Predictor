```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
categorical_columns=df.select_dtypes(include='object').columns
categorical_columns
for col in categorical_columns:
    df[col]=le.fit_transform(df[col])

print(df.head())
    
```

```python
# Splitting Independent Variable
X=df.drop('Attrition',axis=1).values
y=df['Attrition'].values
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.fit_transform(X_test)
```

## Random Forest Classifier

```python
# Apply Random Forest Classifier Algorithm
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=300,random_state=0)
rfc.fit(X_train,y_train)
```

```python
# Test the  model
yf_pred=rfc.predict(X_test)
y
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix,accuracy_score
cmf=confusion_matrix(y_test,yf_pred)
cmf
```

```python
print(f"Accuracy of model: {accuracy_score(y_test,yf_pred)*100}%")
```

## Logistic Regression Algorithm

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
# Test the model
yl_pred=lr.predict(X_test)
```

```python
# confusion merics
from sklearn.metrics import confusion_matrix,accuracy_score
cml=confusion_matrix(yl_pred,y_test)
cml
```

```python
print(f'Accuracy score of model : {accuracy_score(y_test,yl_pred)*100}%')
```

### Decission Tree Algorith

```python
# Apply Decision Tree Algorithm
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
# Test the Mddel
yd_pred=dtc.predict(X_test)
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix,accuracy_score
cmd=confusion_matrix(y_test,yd_pred)
cmd
```

```python
print(f"Accuracy of model : {accuracy_score(y_test,yd_pred)*100}%")
```

### SVM

```python
# Apply SVM algorithm
from sklearn.svm import SVC
svm=SVC()
svm.fit(X_train,y_train)
```

```python
# Test the model
ys_pred=svm.predict(X_test)
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix,accuracy_score
cms=confusion_matrix(y_test,ys_pred)
cms
```

```python
print(f'Accurecy of model : {accuracy_score(y_test,ys_pred)*100}%')
```

# KNN algorithm

```python
# Apply KNN algorithm
from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier(n_neighbors=3)
knc.fit(X_train,y_train)
```

```python
yk_pred=knc.predict(X_test)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix , accuracy_score
cmk=confusion_matrix(y_test,yk_pred)
cmk
```

```python
print(f'Accuracy score of model: {accuracy_score(y_test,yk_pred)*100}%')
```
